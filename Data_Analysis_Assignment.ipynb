````python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})

travel_df = pd.read_csv("travel-times.csv")
travel_df
````

````python
#1 Print the shape of the dataset

travel_df.shape
````

````python
#2 Print the first 15 rows of the dataset

travel_df.head(15)
````

````python
#3 Get information for the features (columns) with missing values

travel_df.info()
````

````python
#4 Drop duplicate values (if any) by keeping only the last instance

travel_df.drop_duplicates(inplace=True, keep = 'last')
travel_df
````

````python
#5 Calculate the total number of missing values (if any) on each column

travel_df.isnull().sum()
````

````python
#6 Create two copies of the dataframe, and then:

travel_df1 = travel_df.copy()
travel_df2 = travel_df.copy()

# drop rows with missing values from the 1st copy (1 points)

travel_df1.dropna()

# drop columns with missing values from the 2nd copy (1 points)

travel_df2.dropna(axis=1)
````

````python
#7 Get summary statistics and see the correlation between the numerical columns

travel_df1.describe()
travel_df3 = travel_df1.describe()

print(travel_df3.corr())
````

````python
#8 Show rows 11 to 14

sub = travel_df.iloc[11:15]
sub

# alternate if index 0 is considered

sub1 = travel_df.iloc[10:14]
sub1
````

````python
#9 Create a subset with trips occurred on November 23, 2011 and January 6, 2012

subset1 = travel_df[((travel_df['Date'] == '11/23/2011') | (travel_df['Date'] == '1/6/2012'))]
subset1
````
